```python
# Transformations
df = df.dropna(how='any',subset=['Embarked'])
y = df['Survived']
df['CabinTF'] = df['Cabin'].empty
df['CabinTF'] = df['CabinTF'].astype(int)
df['Doctor'] = df['Name'].str.contains("Dr.").astype(int)
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 12, 18, 25, 64,999], include_lowest=True, labels=['child', 'teenager', 'young adult', 'adult','old'])
df['Age_Group'] = df['Age_Group'].fillna("adult")

df = df.drop(columns=['PassengerId','Name','Age','Ticket','Survived','Cabin'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(df.select_dtypes(include=[np.number]).columns)

cat_attribs = list(df.select_dtypes(exclude=[np.number]).columns)

one_hot_cat = ['Embarked','Age_Group']

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
         ('std_scaler', MinMaxScaler()),
])


full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat2", OneHotEncoder(sparse=False), one_hot_cat),
        ("cat3", OrdinalEncoder(), ['Sex']),
])
```

```python


X_train = full_pipeline.fit_transform(X_train)
X_test = full_pipeline.fit_transform(X_test)
```
